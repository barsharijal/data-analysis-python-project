```python
def get_pay_with_more_inputs(num_hours, hourly_wage,tax_bracket):
  # pre_tax pay
  pay_pretax = num_hours * hourly_wage
  # after pay_pretax
  pay_pretax_after_tax = pay_pretax * (1 - tax_bracket)
  # pay after tax
  pay_after_tax = pay_pretax_after_tax
  return pay_after_tax
```
